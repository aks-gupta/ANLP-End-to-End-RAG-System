### Pittsburgh Symphony Orchestra

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from requests.packages.urllib3.exceptions import InsecureRequestWarning

requests.packages.urllib3.disable_warnings(InsecureRequestWarning)
```

```python
event_data = []
base_link = "https://www.pittsburghsymphony.org/"
for i in range(1,6):
    driver = webdriver.Chrome()
    if i == 1:
        driver.get(base_link + "calendar")
    else:
        driver.get(base_link + f"calendar?page={i}")
    # Get the page source after interactions
    page_source = driver.page_source
    # Parse the page source with Beautiful Soup
    soup = BeautifulSoup(page_source, "html.parser")
    # Extract data using Beautiful Soup methods
    events = soup.find_all("div", class_="event-details-wrap")
    for event in events:
        event_info = {}
    
        # Extract title
        title = event.find('h3')
        if title:
            event_info['title'] = title.text.strip()
    
        sub_link = event.find('a')
        if sub_link:
            event_info['sub_link'] = base_link + sub_link['href']
    
        time  = event.find('time')
        if time:
            event_info['time'] = time.text.strip()
        venue  = event.find('div', class_="venue")
        if venue:
            event_info['venue'] = venue.text.strip()
        organization  = event.find('div', class_="organization")
        if organization:
            event_info['organization'] = organization.text.strip()
        
            
        event_data.append(event_info)
    driver.quit()
```

```python
with open('data/symphony.txt', 'w') as f:
    for event in event_data:
        f.write(f"Title: {event.get('title', 'N/A')}\n")
        f.write(f"Link: {event.get('sub_link', 'N/A')}\n")
        f.write(f"Time: {event.get('time', 'N/A')}\n")
        f.write(f"Venue: {event.get('venue', 'N/A')}\n")
        f.write(f"Organization: {event.get('organization', 'N/A')}\n")
        f.write("\n")
```

```python
history_data = []
history_links = ["pso_home/web/about-landing/history/history-of-the-pittsburgh-symphony-orchestra","pso_home/web/about-landing/history/history-of-heinz-hall", "pso_home/biographies/pso-conductors/honeck-manfred"]
for link in history_links:
    driver = webdriver.Chrome()
    driver.get(base_link + link)
    page_source = driver.page_source
    # Parse the page source with Beautiful Soup
    soup = BeautifulSoup(page_source, "html.parser")
    # Extract data using Beautiful Soup methods
    try:
        history = soup.find_all("article", class_="bio")
        for para in history:
            history_info = {}
        
            # Extract title
            years = para.find('h2')
            if years:
                history_info['years'] = years.text.strip()
        
            subtitle = para.find('h3')
            if subtitle:
                history_info['subtitle'] = subtitle.text.strip()
        
            desc  = para.find_all('p')
            if desc:
                history_info['description'] = " ".join([d.text.strip() for d in desc])
            
            history_data.append(history_info)
        others = soup.find_all("section", class_="internal-page")
        for para in others:
             data = para.find_all(['h2', 'h3', 'h4', 'p'])
             if data:
                 history_data.append({'data': " ".join([d.text.strip() for d in data if d])})
    except:
        pass
    driver.quit()
```

```python
with open("data/symphony.txt", 'a') as f:
    f.write("\n\nHistory of Pittsburgh Symphony\n")
    for item in history_data:
        if 'years' in item:
            f.write(f"\n{item['years']}\n")
        if 'subtitle' in item:
            f.write(f"{item['subtitle']}\n")
        if 'description' in item:
            sentences = item['description'].split('.')
            for sentence in sentences:
                if sentence.strip():
                    f.write(f"{sentence.strip()}.\n")
        if 'data' in item:
            sentences = item['data'].split('.')
            for sentence in sentences:
                if sentence.strip():
                    f.write(f"{sentence.strip()}.\n")
        f.write("\n")
```

```python
csc = []
base_links = ["https://carnegiesciencecenter.org/calendar/list/?tribe-bar-date=2024-10-26","https://carnegiesciencecenter.org/things-to-do-in-pittsburgh/", "https://carnegiesciencecenter.org/visit/museum-hours/", "https://carnegiesciencecenter.org/visit/pricing/"]

for link in base_links:
    driver = webdriver.Chrome()
    driver.get(link)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    events = soup.find_all("div", class_="entry-content")
    for event in events:
        data = event.find_all(['h2', 'h3', 'h4', 'p', 'time', 'div', 'a'])
        if data:
            csc.append(" ".join([d.text.strip() for d in data if d]))
    driver.quit()
```

```python
import re

cleaned_csc = []
for event in csc:
    cleaned_event = ' '.join(event.split())
    cleaned_event = re.sub(r'(Subscribe to calendar|Google Calendar|iCalendar|Outlook 365|Outlook Live|Export .ics file|Export Outlook .ics file)', '', cleaned_event)
    cleaned_event = cleaned_event.strip()
    if cleaned_event:
        cleaned_csc.append(cleaned_event)

with open('data/carnegie_museums.txt', 'a') as f:
    f.write("\n\nCARNEGIE SCIENCE CENTRE\n")
    for event in cleaned_csc:
        event_parts = event.split('Event Series')
        for part in event_parts:
            part = part.strip()
            if part:
                date_time, *description = part.split(' – ', 1)
                f.write(f"Date/Time: {date_time}\n")
                if description:
                    f.write(f"Description: {description[0]}\n")
                f.write("\n")

cleaned_csc
```

```python
sublinks = []
base_link = "https://en.wikipedia.org/wiki/List_of_museums_in_Pittsburgh"
driver = webdriver.Chrome()
driver.get(base_link)
# Get the page source after interactions
page_source = driver.page_source
# Parse the page source with Beautiful Soup
soup = BeautifulSoup(page_source, "html.parser")
# Extract data using Beautiful Soup methods
links = soup.find("table").find_all('a')
for link in links:
    if link['href'].startswith('/wiki/'):
        sublinks.append(link['href'])
driver.quit()
```

```python
museums = []
base_link = "https://en.wikipedia.org"
for link in sublinks:
    driver = webdriver.Chrome()
    driver.get(base_link+link)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    main_content = soup.find('div', class_ = 'mw-parser-output')
    if main_content:
        for element in main_content(['script', 'style', 'sup', 'table', 'div', 'img']):
            element.decompose()
        text_content = main_content.get_text(separator=' ', strip=True)
        museums.append(text_content)
    else:
        print('Main content area not found.')
    
```

```python
cleaned_museums = []
for museum in museums:
    cleaned_text = ' '.join(museum.split())
    cleaned_museums.append(cleaned_text)

with open("data/museums_list.txt", 'w', encoding='utf-8') as f:
    for museum_info in cleaned_museums:
        f.write(museum_info + '\n\n')
```

#### Pittsburgh Festivals

```python
import time


sublinks = []
fests = []
base_link = "https://www.visitpittsburgh.com/events-festivals/"
for i in range(1,21):
    driver = webdriver.Chrome()
    if i == 1:
        driver.get(base_link)
    else:
        driver.get(base_link + f"?page={i}")
    # Scroll the page to load all the events
    last_height = driver.execute_script("return document.body.scrollHeight")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(2)  # Wait for the page to load
        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height
    # Get the page source after interactions
    page_source = driver.page_source
    # Parse the page source with Beautiful Soup
    soup = BeautifulSoup(page_source, "html.parser")
    # Extract data using Beautiful Soup methods
    events = soup.find_all('div', class_='card')
    for event in events:
        details = event.find('div', class_= "card__main").find_all(['h2', 'h3', 'h4', 'p', 'span'])
        if details:
            fests.append(" ".join([d.text.strip() for d in details if d]))
        link = event.find('a')
        if link['href'].startswith('/events/'):
            sublinks.append(link['href'])
    driver.quit()
```

```python
fest_pages = ""
base_link = "https://www.visitpittsburgh.com"
for link in sublinks:
    try:
        driver = webdriver.Chrome()
        driver.get(base_link+link)
        page_source = driver.page_source
        soup = BeautifulSoup(page_source, "html.parser")
        date = soup.find('h2', class_ = "detail__dateheading").text.strip()
        fest_pages += date
        header = soup.find('h1', class_ = "detail__heading").text.strip()
        main_content = soup.find('div', class_ = 'detail__summary').find_all(['h2', 'h3', 'h4', 'p', 'span'])
        fest_pages += header
        if main_content:
            fest_pages += " ".join([d.text.strip() for d in main_content if d])
    except:
        print("skipped")
    driver.quit()
```

```python
with open("data/festivals.txt", "a") as f:
    lines = fest_pages.split()
    for i in range(0, len(lines), 20):
        chunk = " ".join(lines[i:i+20])
        f.write(chunk + "\n")
```

```python
wikis = []
base_links = ["https://en.wikipedia.org/wiki/Pittsburgh_Symphony_Orchestra", "https://en.wikipedia.org/wiki/Pittsburgh_Opera", "https://trustarts.org/pct_home/about/history", "https://trustarts.org/pct_home/about/cultural-district"]

for link in base_links:
    driver = webdriver.Chrome()
    driver.get(link)
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    content = soup.find("main")
    data = content.find_all(['h1', 'h2', 'h3', 'h4', 'p'])
    if data:
        wikis.append(" ".join([d.text.strip() for d in data if d]))
    driver.quit()
```
